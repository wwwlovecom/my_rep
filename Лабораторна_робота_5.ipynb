```python
import plotly.express as px
```

Завдання 1.

```python
df = px.data.iris()
df.head()
```

Графік 1

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    size='petal_length',
    hover_data=['petal_width'],
    color_discrete_map={
        "setosa": "#FF5733",
        "versicolor": "#33FF57",
        "virginica": "#3357FF"
    }
)

fig.update_layout(
    plot_bgcolor="#1e1e1e",
    paper_bgcolor="#111111",
    font=dict(color="white")
)

fig.show()
```

Графік 2

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="petal_length",
    color_continuous_scale="Turbo"
)
fig.show()
```

Графік 3

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", symbol="species")
fig.show()
```

Завдання 3

```python
df = px.data.gapminder().query("continent == 'Oceania'")
df.head()
```

```python
fig = px.line(df, x='lifeExp', y='gdpPercap', color='country')
fig.show()
```

```python
df = px.data.gapminder().query("year == 2007")

fig = px.scatter(
    df, x="gdpPercap", y="lifeExp",
    size="pop", color="continent",
    hover_name="country", log_x=True,
    title="GDP vs Life Expectancy (2007)",
    size_max=60
)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

N = 1000

#1
t = np.linspace(0, 5*np.pi, 50)
y = 2*np.cos(t)*(t-np.pi/4)

fig = go.Figure(data=go.Scatter(x=t, y=y, mode='markers'))
fig.show()
```

```python
#2
t2 = np.linspace(0, 2*np.pi, 100)
y2 = 0.5*np.sin(2*t2+np.pi/3)

fig2 = go.Figure(data=go.Scatter(x=t2, y=y2, mode='markers'))
fig2.show()
```

```python
#3
t3 = np.linspace(-2*np.pi, 2*np.pi, 150)
y3 = 2*np.tan(0.5*t3+np.pi/6)

fig3 = go.Figure(data=go.Scatter(x=t3, y=y3, mode='markers'))
fig3.show()
```
